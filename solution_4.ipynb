## 2. Write a function to find moving average in an array over a window:
## Test it over [3, 5, 7, 2, 8, 10, 11, 65, 72, 81, 99, 100, 150] and window of 3

```python
#Function to calculate moving average....
def mov_avg_cal():
    x=np.array([3, 5, 7, 2, 8, 10, 11, 65, 72, 81, 99, 100, 150])
    k=3
    n=13
    return [sum([pt for pt in x[int(np.where(x==pt)[0]): int(np.where(x==pt)[0])+3]])/3 for pt in x if pt<x[n-k+1]]
mov_avg_cal()
```

```python
import pandas as pd
df = pd.DataFrame({'x':[7,2,0,3,4,2,5,0,3,4]})
se = []
flag = 0
for ct in df.x:
    if ct == 0:
        se.append(0) 
        flag = 1
        cp = 0
    elif flag == 1:
            cp+=1
            se.append(cp)
    else:
        se.append(0)
df = pd.DataFrame({'x':[7,2,0,3,4,2,5,0,3,4],'y':se})
df
```

```python
#creating date range....
per = pd.bdate_range(start='2015-01-01', end='2015-12-31')
#dt will return DatetimeIndex()....
rd = np.random.randint(1,261,261)
per_rd = list(zip(per,rd))
df_one = pd.DataFrame(per_rd, columns=('Datetimeindex','Random_Num'))
fst_qa = df_one[0:86]
sec_qa = df_one[86:173]
lst_qa = df_one[173:261]
#fetching highesh date from each quarter....
a=fst_qa[fst_qa.Random_Num == max(fst_qa.Random_Num)]['Datetimeindex'].values
b=sec_qa[sec_qa.Random_Num == max(sec_qa.Random_Num)]['Datetimeindex'].values
c=lst_qa[lst_qa.Random_Num == max(lst_qa.Random_Num)]['Datetimeindex'].values
dl = pd.DataFrame(data=[a,b,c], index=['First QA','Second QA', 'Last QA'], columns=['Date of highest value'])
dl
```
